```python
import pandas as pd


# --> read df
try  :
    df = pd.read_csv("avito_appartements.csv")
    print("read data successfully")
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# --> nomber de lignes et colon
print(f"Nombre de lignes : {df.shape[0]}")
print(f"Nombre de colonnes : {df.shape[1]}")
```

```python
print("\nAperçu des données :")
print(df.head())
print(df.tail())
```

```python
# --> afficher les columns
print(df.columns.tolist())
```

```python
# --> des info par data
print(df.info())
```

```python
# --> types des colonnes
print(df.dtypes)
```

```python
# --> nomber des null par colonnes
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# --> poursontage par nomber null
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.sort_values(ascending=False))
```

```python
# --> nomber dublicat ( par url )
duplicates_url = df.duplicated(subset=["url"]).sum()
print(f"Duplicates based on URL: {duplicates_url}")
```

```python
# --> min , max , mean , median , std ( price )
if "price" in df.columns:
    print("Min price :", df["price"].min())
    print("Max price :", df["price"].max())
    print("Mean price :", df["price"].mean())
    print("Median price :", df["price"].median())
    print("Price standard deviation :", df["price"].std())
```

```python
# nomber repete des city
if "city" in df.columns:
    print(df["city"].value_counts().head(10))
```

```python
# --> les surface
if "surface" in df.columns:
    print(df["surface"].unique())
```
